```python
from docx import Document
import re
import os

def word_to_latex(doc_path, output_dir):
    # Load the Word document
    doc = Document(doc_path)
    
    # Regex to identify chapter headers
    chapter_regex = re.compile(r'Chapter (\d+):?.*')
    current_chapter = None
    current_content = []
    chapters = []

    # Iterate through paragraphs in the document
    for para in doc.paragraphs:
        text = para.text.strip()
        
        # Check if this paragraph is a new chapter
        match = chapter_regex.match(text)
        if match:
            # Save the current chapter if it exists
            if current_chapter:
                chapters.append((current_chapter, "\n".join(current_content)))
            
            # Start a new chapter
            current_chapter = text  # Full chapter title
            current_content = []
        else:
            # Process paragraph content with formatting
            if current_chapter:
                processed_text = process_text(para)
                current_content.append(processed_text + "\n")
    
    # Add the last chapter after the loop
    if current_chapter:
        chapters.append((current_chapter, "\n".join(current_content)))
    
    # Save each chapter as a standalone LaTeX file
    for title, content in chapters:
        save_to_latex(title, content, output_dir)

def process_text(para):
    # Process text and maintain formatting
    text_content = ""
    # print(para.alignment)
    if para.alignment: 
        return r"""
        \begin{center}
                {\Huge  . \quad . \quad .}
            \end{center}"""
    for run in para.runs:
        # print(run.font.size)
        run_text = escape_latex(run.text)
        if run.italic:
            run_text = f"\\textit{{{run_text}}}"
        text_content += run_text
    return text_content

def escape_latex(text):
    # Escape LaTeX special characters
    special_chars = {
        # '\\': r'\textbackslash{}',
        '&':  r'\&',
        '%':  r'\%',
        '$':  r"\$",
        '#':  r'\#',
        '_':  r'\_',
        '{':  r'\{',
        '}':  r'\}',
        '~':  r'\textasciitilde{}',
        '^':  r'\textasciicircum{}',
    }
    for char, escaped_char in special_chars.items():
        text = text.replace(char, escaped_char)
    return text


def save_to_latex(chapter_title, content, output_dir):
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Extract chapter number for filename
    chapter_number = re.search(r'Chapter\s*(\d+)', chapter_title).group(1)
    file_path = os.path.join(output_dir, f"Chapter_{chapter_number}.tex")
    
    # Add necessary LaTeX structure and format the chapter title and content
    complete_content = (
        # "\\documentclass{book}\n"
        # "\\usepackage[utf8]{inputenc}\n"
        # "\\begin{document}\n" +
        f"\\chapter{{{escape_latex(chapter_title.split(':')[1].strip())}}}\n" +  # Full chapter title
        content 
        # "\\end{document}\n"
    )
    
    # Write content to the .tex file
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(complete_content)
    print(f"{chapter_title} saved to {file_path}")

# Usage
# word_to_latex("your_document.docx", "output_tex_files")
```

```python
import pypandoc
import re
import os

def word_to_latex(doc_path, output_dir):
    # Convert the Word document to plain text
    text_content = pypandoc.convert_file(doc_path, 'plain')
    
    # Regex to find each "Chapter x:" header with optional extra text on the same line
    chapter_regex = re.compile(r'(Chapter \d+[:]?.*?)(?=\nChapter \d+[:]?|$)', re.DOTALL)
    chapters = chapter_regex.findall(text_content)
    
    # Process and save each chapter
    for chapter in chapters:
        # Split chapter title and content based on the first newline after the title
        title, _, content = chapter.partition('\n')
        
        # Save each chapter as a standalone LaTeX file
        save_to_latex(title.strip(), content.strip(), output_dir)

def save_to_latex(chapter_title, content, output_dir):
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Extract chapter number for filename
    chapter_number = re.search(r'Chapter\s*(\d+)', chapter_title).group(1)
    file_path = os.path.join(output_dir, f"Chapter_{chapter_number}.tex")
    
    # Add necessary LaTeX structure and format the chapter title and content
    complete_content = (
        "\\documentclass{book}\n"
        "\\usepackage[utf8]{inputenc}\n"
        "\\begin{document}\n" +
        f"\\chapter{{{chapter_title}}}\n" +  # Full chapter title with colon
        content +
        "\\end{document}\n"
    )
    
    # Write content to the .tex file
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(complete_content)
    print(f"{chapter_title} saved to {file_path}")

# Usage
# word_to_latex("your_document.docx", "output_tex_files")
```
